### Code (cleaned)


````python


import kagglehub
import pandas as pd

# Download dataset
path = kagglehub.dataset_download("laotse/credit-risk-dataset")

# Load dataset
df = pd.read_csv(f"{path}/credit_risk_dataset.csv")

df.head()
````

````python


df.info()
df.describe()
df.shape
````



---

## üìç 4. Data Quality Checks & Cleaning

### Duplicate Handling


````python

df = df.drop_duplicates()
````


### Missing Value Treatment (Explained, not noisy)


````python

# Employment length ‚Üí mean imputation
df['person_emp_length'].fillna(df['person_emp_length'].mean(), inplace=True)

# Interest rate ‚Üí median imputation (robust to outliers)
df['loan_int_rate'].fillna(df['loan_int_rate'].median(), inplace=True)

df.isnull().sum()
````


‚úî Matches **Advanced preprocessing requirement**

---

## üìç 5. Feature Categorization


````python

num_cols = [
    'person_age',
    'person_income',
    'loan_amnt',
    'cb_person_cred_hist_length'
]

cat_cols = [
    'person_home_ownership',
    'loan_intent',
    'loan_grade',
    'cb_person_default_on_file'
]

target = 'loan_status'
````


---

## üìç 7. Feature Encoding (Industry-Correct)


````python

from sklearn.preprocessing import LabelEncoder, OrdinalEncoder

# Nominal encoding
home_encoder = LabelEncoder()
intent_encoder = LabelEncoder()

df['person_home_ownership'] = home_encoder.fit_transform(df['person_home_ownership'])
df['loan_intent'] = intent_encoder.fit_transform(df['loan_intent'])

# Ordinal encoding for loan grade
grade_encoder = OrdinalEncoder(categories=[['A','B','C','D','E','F','G']])
df['loan_grade'] = grade_encoder.fit_transform(df[['loan_grade']])

# Binary mapping
df['default_history'] = df['cb_person_default_on_file'].map({'Y':1, 'N':0})

df.drop(columns=['cb_person_default_on_file'], inplace=True)
````


‚úî Exactly what a **credit-risk ML system expects**

---

## üìç 8. Feature Renaming (Deployment Friendly)


````python

df.rename(columns={
    'person_age': 'age',
    'person_income': 'income',
    'person_emp_length': 'employment_length',
    'loan_amnt': 'loan_amount',
    'loan_int_rate': 'interest_rate',
    'cb_person_cred_hist_length': 'credit_hist_length'
}, inplace=True)
````


---

## üìç 9. Feature Scaling & Train-Test Split


````python

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X = df.drop(columns=['loan_status'])
y = df['loan_status']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42
)
````


---

## üìç 10. Model Building & Comparison


````python

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
import xgboost as xgb
from sklearn.metrics import classification_report, roc_auc_score

models = {
    "Logistic Regression": LogisticRegression(class_weight='balanced', max_iter=10000),
    "Random Forest": RandomForestClassifier(class_weight='balanced', n_estimators=200),
    "Gradient Boosting": GradientBoostingClassifier(),
    "XGBoost": xgb.XGBClassifier(
        scale_pos_weight=len(y[y==0])/len(y[y==1]),
        eval_metric='logloss'
    ),
    "SVM": SVC(class_weight='balanced', probability=True)
}

for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    print(f"\n{name}")
    print(classification_report(y_test, preds))
````


---

## üìç 11. Hyperparameter Tuning (Advanced Level)


````python

from sklearn.model_selection import RandomizedSearchCV

xgb_model = xgb.XGBClassifier(
    objective='binary:logistic',
    eval_metric='logloss',
    random_state=42
)

param_grid = {
    'n_estimators': [200, 400, 600],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.05, 0.1],
    'scale_pos_weight': [len(y[y==0]) / len(y[y==1])]
}

search = RandomizedSearchCV(
    xgb_model,
    param_grid,
    scoring='roc_auc',
    cv=5,
    n_iter=30,
    n_jobs=-1
)

search.fit(X_train, y_train)
best_model = search.best_estimator_
````


---

## üìç 12. Threshold Tuning & Evaluation


````python

y_prob = best_model.predict_proba(X_test)[:,1]

threshold = 0.6
y_pred = (y_prob >= threshold).astype(int)

print(classification_report(y_test, y_pred))
````


---

## üìç 13. Explainability (SHAP)


````python

import shap

explainer = shap.TreeExplainer(best_model)
shap_values = explainer.shap_values(X_test)

shap.summary_plot(shap_values, X_test)
````


---

## üìç 14. Model Persistence (Deployment Ready)


````python

import joblib

joblib.dump({
    "model": best_model,
    "scaler": scaler,
    "home_encoder": home_encoder,
    "intent_encoder": intent_encoder,
    "grade_encoder": grade_encoder,
    "threshold": threshold
}, "credit_risk_model.pkl")
````
